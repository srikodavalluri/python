 b) Write a function dup to find all duplicates in the list.

```python
def dup(list):
  temp=[]
  for i in list:
    c=list.count(i)
    if (c>1 and (i not in temp)):
      temp.append(i)
  print("Duplicates in the list:",temp) 
 
ls=[44,20,10,15,44,17,15]
print("Given elements in the list:",ls)
dup(ls)
```
